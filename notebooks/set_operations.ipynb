```python
# You can combine two lists with +
A = [1, 2, 3]
B = [4, 5, 6]
C = A + B
print(f"A: {A}, B: {B}, C: {C}")

# You can also use list.extend() but that extends the list in place and does not return anything.
# List.extend() adds the elements of the second list one by one.
D = A.copy()  # Need to copy A, otherwise mutating D will also mutate A.
D.extend(B)
print(f"D: {D}")

# You can use list.append() but that adds a list as one element.
# It also extends the list in place and does not return anything.
F = A.copy()
F.append(B)
print(f"F: {F}")
```

## Lists, tuples, sets

```python
A = [0, 1, 2, 3, 2, 1]  # List
B = (3, 2, 2, 1, 1, 0)  # Tuple
C = {0, 1, 2, 3, 2, 1}  # Set
print(f"A: {A}, B: {B}, C: {C}")
print(f"|A| = {len(A)}, |B| = {len(B)}, |C| = {len(C)}")
print(f"A has {len(set(A))} unique elements")
print(f"B as a set = {set(B)}")
```
