## Music equivalence ops

```python
from music21 import note
```

```python
N = note.Note
```

```python
N('C4').show()
```

```python
c = N('C4')
c2 = N('C4')
c == c2
```

```python
c3 = N('C5')
```

```python
c == c3
```

```python
c2.duration.quarterLength = 4.0
```

```python
c == c2
```

```python
c4 = N('C4')
c == c4
```

```python
c.lyric = 'hello'
c.show()
```

```python
c4.lyric = 'bonjour'
c == c4
```

```python
bs = N('B#4')
bs.show()
```

```python
bs.pitch.octave = 3
bs.show()
```

```python
c == bs
```

```python
def new_eq(a, b):
    return a.pitch.midi == b.pitch.midi
```

```python
N.__eq__ = new_eq
```

```python
eis = N('E#4')
f = N('F4')
eis == f
```
